```python
df['production_companies'] = df['production_companies'].fillna('')
df['production_countries'] = df['production_countries'].fillna('')
df['homepage'] = df['homepage'].fillna('')
df['tagline'] = df['tagline'].fillna('')
df['overview'] = df['overview'].fillna('')
```

```python
df['poster_path'] = df['poster_path'].fillna('https://cdn.pixabay.com/photo/2016/04/14/07/50/film-1328405_1280.jpg')
df
```

```python
df['created_by']= df['created_by'].apply(lambda x:x.replace(' ',''))
```

```python
df['production_companies']= df['production_companies'].apply(lambda x:x.replace(' ',''))
df['production_countries']= df['production_countries'].apply(lambda x:x.replace(' ',''))
df
```

```python
df['description']=df['overview']+' '+df['created_by']+' '+df['production_companies']+' '+df['production_countries']+' '+df['genres']+' '+df['type'].apply(lambda x:x.replace(' ',''))+df['tagline'].apply(lambda x:x.replace(' ',''))
```

```python
df2 = df[['name','poster_path','description','number_of_seasons','number_of_episodes','popularity','homepage','type','tagline']]
df2.describe()
```

```python
df2['description']=df2['description'].str.lower()
```

```python
df2
```

```python
import re
def name_checker(name):
    return re.sub('[^A-Za-z0-9 ]','',name)
df2['name']= df2['name'].apply(name_checker)
```

```python
df3 = df2[~(df2['name'] == '')].reset_index(drop=True)
df3
```

```python
df3['description']= df3['description'].apply(name_checker)
df3
```

```python
import numpy as np 
df3['description']=df3['description'].apply(lambda x:x if len(str(x).split())>=20 else np.nan)
```

```python
df3.isnull().sum()
```

```python
df4  = df3.dropna(how='any').reset_index(drop=True)
df4
```

```python
df4 = df4.set_index('name')
```

```python
df4
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=200,stop_words='english')
vectors = cv.fit_transform(df4['description'])
vectors.shape
```

```python
vectors
```

```python
cols = cv.get_feature_names_out()
```

```python
vectors = vectors.toarray()
```

```python
vectors
```

```python
new_df= pd.DataFrame(data=vectors,index=df4.index,columns=cols)
new_df
```

```python
new_df['poster_path']=df4['poster_path']
new_df['number_of_seasons']=df4['number_of_seasons']
new_df['number_of_episodes']=df4['number_of_episodes']
new_df['popularity']=df4['popularity']
new_df['homepage']=df4['homepage']
new_df['tagline']=df4['tagline']
new_df['type']=df4['type']
new_df
```

```python
new_df
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric='euclidean',algorithm='brute')
model.fit(new_df.drop(columns=['poster_path','number_of_seasons','number_of_episodes','popularity','homepage','type','tagline'],axis=1))
train_df = new_df.drop(columns=['poster_path','number_of_seasons','number_of_episodes','popularity','homepage','type','tagline'],axis=1)
```

```python
series_name = 'Stranger Things'
distances, indices = model.kneighbors(X=train_df.loc[series_name,:].values.reshape(1,-1),
                                           n_neighbors=6)
```
